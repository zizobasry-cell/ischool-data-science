```python
import mathassistant1
x= int(input("Enter your first number"))
y= int(input("Enter your second number"))

while True:
  opt = (input("enter your option : " ))
  if opt == "min" :
    mathassistant1.mininumber(x , y )
  elif opt == "max" :
    mathassistant1.maximumnumber(x, y)
  elif opt == "add" :
    mathassistant1.add(x , y)
  elif opt == "subtraction" :
    mathassistant1.Subtractionnumber(x,y)
  elif opt == "multiplication":
    mathassistant1.multiplicationnumber(x,y)
  elif opt == "division" :
    mathassistant1.divisionnumber(x , y)
  elif opt == "exit" :
    print("Goodbye")
  else :
    print("invalid operation")
```
