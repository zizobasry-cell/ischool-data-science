```python
items_prices={
"shipsy" : 10 , "cola" : 15 , "apple" : 15 , "headset" : 100 , "sugar" : 20
}

cart = []
while True :
  order=(input(" enter your order : "))
  if order == "add item" :
    item =(input("""    shipsy : 10    cola : 15     apple : 15     headset: 100   sugar: 20  Enter  your item   """))
    cart.append(item)
    print(cart)
  elif order == "remove item":
    print(cart)
    removed_item=(input(" which item to delete :  "))
    cart.remove(removed_item)
    print(cart)
  elif order == "checkout":
    total_price = 0
    for item in cart:
      total_price += items_prices[item]
    print(f" total ammount: {total_price}")
  elif order == "exit" :
    items_prices.clear()
    print(items_prices)

```
