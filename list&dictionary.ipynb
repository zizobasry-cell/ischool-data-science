```python
to_do_list=["go to the gym" , "call mum" , "do homework"]
print(to_do_list)
```

```python
to_do_list.insert(2 , "eat lunch")
print(to_do_list)
```

```python
for i in range(len(to_do_list)):
  print(to_do_list[i])
```
