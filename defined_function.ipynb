```python
def max(x,y):
  if x>y:
    print(x)
  else:
    print(y)
```

```python
x= int(input("Enter your first number"))
y= int(input("Enter your second number"))
while True:
  opt=input("Enter your operation : ")
  if opt == "min":
    min(x,y)
  elif opt == "max" :
    max(x,y)
  elif opt == "exit":
    break
  else:
    print("Invalid operation")
```
